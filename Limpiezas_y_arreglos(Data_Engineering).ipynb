```python
import pandas as pd
import numpy as np

df_amazon = pd.read_csv(r"MLOpsReviews\amazon_prime_titles.csv")
df_disney = pd.read_csv(r"MLOpsReviews\disney_plus_titles.csv")
df_hulu = pd.read_csv(r"MLOpsReviews\hulu_titles.csv")
df_netflix = pd.read_csv(r"MLOpsReviews\netflix_titles.csv")
```

```python
#Function to add the initials to the beggining of  "show_id"
def adding_initials (df, letter):
    df["show_id"] = df["show_id"].apply(lambda x: letter + str(x))
```

```python
#Function to change ratings to g
def ratings_nan_to_g (df):
    df["rating"].fillna("g", inplace=True)
```

```python
ratings_nan_to_g(df_amazon)
df_amazon.head()
```

```python
ratings_nan_to_g(df_hulu)
ratings_nan_to_g(df_disney)
ratings_nan_to_g(df_netflix)
```

```python
#At this moment, I have realized that I can merge the datasets since they share the same column names
#It facilitates the conversion without the need to repeat for each dataset, but in the end I will separate them again.
df_merged = pd.concat([df_amazon, df_disney, df_hulu, df_netflix], ignore_index=True)
df_merged["show_id"]
```

```python
#Cleaning for future datetimes conversions
df_merged["date_added"] = df_merged["date_added"].str.replace(",", " ")
df_merged.head()
```

```python
#Changing format to "AAAA-mm-dd"
df_merged["date_added"] = pd.to_datetime(df_merged["date_added"])
df_merged.head()
```

```python
#changing strings to lowercase
obj_cols = df_merged.select_dtypes(include="object")
df_merged[obj_cols.columns] = obj_cols.applymap(lambda x: x.lower() if isinstance(x, str) else x)
df_merged.head()
```

```python
#Separating the 'duration' column into two columns, 'duration_int' and 'duration_type'. For my own preference, I want the new columns to be in the same position as the old one.
duration_separado = df_merged['duration'].str.extract(r'(\d+)\s*(\S+)')
duration_separado.columns = ["duration_int", "duration_type"]
indice_col= df_merged.columns.get_loc("duration")
df_merged = df_merged.iloc[:, :indice_col+1].join(duration_separado).join((df_merged.iloc[:, indice_col+2:]))
df_merged["duration_type"].astype(str)
df_merged.head()
```

```python
#Here I am separating the datasets. 
#Using 'groupby', I take the first character in the 'show_id' column, which is the initial letter of the company name, and then I assign it to a dictionary.
dfs = {}
groups = df_merged.groupby(df_merged["show_id"].str[0])
for name, group in groups:
    dfs[name] = group
```

```python
#Just to see the before and after
#BEFORE
df_amazon.head()
```

```python
#AFTER
df_amazon = dfs["a"]
df_amazon.head()
```

```python
df_disney = dfs["d"]
df_hulu = dfs["h"]
df_netflix = dfs["n"]
```
